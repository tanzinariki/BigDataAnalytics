```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df.drop(columns=['Churn', 'CustomerID'])
y = df['Churn']

# Convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)

print("Class distribution before split:\n", y.value_counts())

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("\nTraining class size:", X_train.shape)
print("Test class size:", X_test.shape)

# Check class distribution in both sets
print("\nTraining class distribution:\n", y_train.value_counts())
print()  # Print new line for better readability
print("Test class distribution:\n", y_test.value_counts())
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42, class_weight="balanced")

# Train the model
dt_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Predictions
y_pred = dt_classifier.predict(X_test)

# Evaluate Decision Tree model
print("\nDecision Tree Performance before hyperparameter tuning:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision: {precision_score(y_test, y_pred):.3f}")
print(f"Recall: {recall_score(y_test, y_pred):.3f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.3f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5, scoring='f1')

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Best model
best_dt_classifier = grid_search.best_estimator_
```

```python
# Predict on the test set
y_pred = best_dt_classifier.predict(X_test)

# Evaluate the model
print("\nDecision Tree Performance after hyperparameter tuning:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision: {precision_score(y_test, y_pred):.3f}")
print(f"Recall: {recall_score(y_test, y_pred):.3f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.3f}")

# Confusion Matrix for Decision Tree 
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix for Decision Tree:")
print(conf_matrix)

# Plot Confusion Matrix using Seaborn
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Greens", xticklabels=["No Churn", "Churn"], yticklabels=["No Churn", "Churn"])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix for Decision Tree')
plt.show()
```

```python
from sklearn.tree import plot_tree

# Plot the decision tree
plt.figure(figsize=(30, 20))
plot_tree(best_dt_classifier, filled=True, feature_names=X.columns, class_names=['No Churn', 'Churn'], rounded=True, fontsize=12, max_depth=3)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42, class_weight="balanced")

# Train the model
rf_classifier.fit(X_train, y_train)
# Predictions
y_pred_rf = rf_classifier.predict(X_test)

# Evaluate Random Forest model
print("\nRandom Forest Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.3f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.3f}")
print(f"Recall: {recall_score(y_test, y_pred_rf):.3f}")
print(f"F1 Score: {f1_score(y_test, y_pred_rf):.3f}")
```

```python
from sklearn.model_selection import GridSearchCV
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search_rf = GridSearchCV(estimator=rf_classifier, param_grid=param_grid_rf, cv=5, scoring='f1')
grid_search_rf.fit(X_train, y_train)

# Output best combination of hyperparameters found
print("Best Random Forest Parameters:", grid_search_rf.best_params_)
```

```python
# Predict on the test set
y_pred_rf = grid_search_rf.predict(X_test)

# Evaluate the model
# Evaluate Random Forest model
print("\nRandom Forest Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.3f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.3f}")
print(f"Recall: {recall_score(y_test, y_pred_rf):.3f}")
print(f"F1 Score: {f1_score(y_test, y_pred_rf):.3f}")
```

```python
# Confusion Matrix for Random Forest after hyperparameter tuning
conf_matrix = confusion_matrix(y_test, y_pred_rf)
print("\nConfusion Matrix for Random Forest:")
print(conf_matrix)

# Plot Confusion Matrix using Seaborn
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Greens", xticklabels=["No Churn", "Churn"], yticklabels=["No Churn", "Churn"])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix Plot for Random Forest')
plt.show()
```
